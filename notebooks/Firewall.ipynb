```python
from peft import PeftModel
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
base_model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-uncased", num_labels=2)
model = PeftModel.from_pretrained(base_model, "kunjcr2/bert-lora")
model = model.merge_and_unload()  # optional: merge adapter into base weights permanently
```

```python
input_text = ["what is the integral of x^2 from 0 to 1"]
# Added return_tensors='pt' to fix the AttributeError
res = model(**tokenizer(input_text, return_tensors="pt", padding=True))
res
```
